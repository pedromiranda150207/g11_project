```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
df = pd.read_csv(r"C:\Users\Isabe\Downloads\g11_Delivery_warehouses.csv", sep=";", parse_dates=['order_date','created_date'], dayfirst=True)
```

```python
df.columns = df.columns.str.strip()
```

```python
list(df.columns)
```

```python
df_vehicle = df.loc[:, ["vehicle_id", "type"]]
```

```python
df_vehicle = df_vehicle.dropna(subset=["vehicle_type", "fuel_type"])
```

```python
df_vehicle = df_vehicle.drop_duplicates()
```

```python
df_vehicle.insert(0, "id", range(1, len(df_vehicle) + 1))
```

```python
df_vehicle["id"] = df_vehicle["id"].astype(int)
```

```python
df_vehicle.head(3)
```

```python
engine = create_engine("sqlite:///DATABASEFINAL.db")
```

```python
df_vehicle.to_sql("vehicle", con=engine, if_exists="replace", index=False)
```

```python
result = pd.read_sql('SELECT * FROM vehicle', con=engine)
```

```python
result.head(3)
```

```python
df_vehicle.to_csv("vehicle.csv", sep=";", index=False)
```

```python
engine.dispose()
```

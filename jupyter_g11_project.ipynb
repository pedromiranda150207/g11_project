```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
df = pd.read_csv("g11_Delivery_warehouses.csv", sep = ';', parse_dates=['order_date','created_date'], dayfirst=True)
```

```python
df_warehouse = df.loc[:, ['warehouse_id', 'designation', 'type']]
```

```python
df_warehouse = df_warehouse.dropna(subset=['warehouse_id', 'designation', 'type'])
```

```python
df_warehouse = df_warehouse.drop_duplicates(subset='warehouse_id', keep='first')
```

```python
df_warehouse['warehouse_id'] = df_warehouse['warehouse_id'].astype(int)
```

```python
df_warehouse.rename(columns={'warehouse_id': 'id'}, inplace=True)
```

```python
df_warehouse.head(3)
```

```python
engine = create_engine('sqlite:///DATABASEFINAL.db')
```

```python
df_warehouse.to_sql('warehouse', con=engine, if_exists='replace', index=False)
```

```python
result = pd.read_sql('SELECT * FROM Warehouse', con=engine)
```

```python
result.head(3)
```

```python
df_warehouse.to_csv('warehouse.csv', sep=';', index=False)
```

```python
engine.dispose()
```

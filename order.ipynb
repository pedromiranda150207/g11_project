```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
df = pd.read_csv(r"C:\Users\Isabe\Downloads\g11_Delivery_warehouses.csv", sep=";", parse_dates=['order_date','created_date'], dayfirst=True)
```

```python
df_order = df.loc[:, ["cost", "order_date"]]
```

```python
df_order = df_order.dropna(subset=["cost", "order_date"])
```

```python
df_order.insert(0, "id", range(1, len(df_order) + 1))
```

```python
df_order["id"] = df_order["id"].astype(int)
```

```python
df_order["cost"] = df_order["cost"].astype(int)
```

```python
df_order.head(3)
```

```python
engine = create_engine('sqlite:///DATABASEFINAL.db')
```

```python
df_order.to_sql('order', con=engine, if_exists='replace', index=False)
```

```python
result = pd.read_sql('SELECT * FROM "order"', con=engine)
```
